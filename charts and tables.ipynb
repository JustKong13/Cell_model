```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
import pandas as pd 
```

```python
# Density plot combined

original_concentrations = pd.read_csv("metabolite_concentrations.csv")
additional_concentrations = pd.read_csv("additional_concentrations.csv")


datasets = [original_concentrations, additional_concentrations]

i = 0
for subset in datasets:
    i = i + 1
    if i == 1:
        label = "Original Biomass Makeup"
        color = "blue"
    else:
        label = "Improved Biomass Makeup"
        color = "orange"

    splot = sns.distplot(np.log10(abs(subset["Metabolic Weight (g)"])), hist = False, kde = True,
                 kde_kws = {'linewidth': 0.5, "bw_adjust": 0.1}, label=label, color=color, bins= 5000)
    

plt.xlim([-6,0])
plt.legend(prop={'size': 8}, title = 'Model')
plt.title("Density plot with Original and Improved Biomass makeup")
```

```python
# Density plot combined

original_concentrations = pd.read_csv("metabolite_concentrations.csv")
additional_concentrations = pd.read_csv("additional_concentrations.csv")


datasets = [original_concentrations, additional_concentrations]

i = 0
for subset in datasets:
    i = i + 1
    if i == 1:
        label = "Original Biomass Makeup"
        color = "blue"
    else:
        label = "Improved Biomass Makeup"
        color = "orange"

    splot = sns.distplot((abs(subset["Log Weight"])), hist = False, kde = True,
                 kde_kws = {'linewidth': 0.5, "bw_adjust": 0.1}, label=label, color=color, bins= 5000000000)
    
    


plt.xlim(7,0)
plt.legend(prop={'size': 8}, title = 'Model')
plt.title("Density plot with Original and Improved Biomass makeup")
```

```python
# Pie chart of the total mass of the original metabolites vs. total mass of the new biomass

labels = 'Original Biomass', 'Added biomass'
sizes = [0.99, 0.99-0.97]
colors = ['gold', 'yellowgreen']

# Plot
plt.title("Breakdown of improved biomass")
plt.pie(sizes, labels=labels, colors=colors,
autopct='%1.1f%%', startangle=90)

plt.axis('equal')
plt.show()
```
